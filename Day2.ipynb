# Control Flow Statement

```python
Control flow statements:
    1. conditional statements [If, elif else/terniary if]
    2. loop [for (ranged loop) while (conditional loop)]
    
While loop: 
when u don’t know how many time 

For loop:
when you know how many times you want to iterate
```

```python
 1. conditional statements [If, elif else/terniary if] (indexing)
    
There are tow type while writing codes
#scenario 1
when we write codes there is a way to write and we are want to take some decision and based on that we dont eant to execute 
l5 and l6. Hence there will be many conditions and so we write 'contitional statements'-- eg,

l1
l2
l3
l4

l5

l6
```

```python
if statement syntax:
#syntax 1 (we can have n number of elif statement)
if condition/s:
    line1...
    line2..
    
elif condition/s:
    l1
    l2
    
elif condition/s:
    l1
    l2
    
else:
    line 3..
    line 4..
when the code is run it checks for the 1st condition and if the first condition is true then it will not even check the 
2nd condition...it will check the 2nd condition only when the 1st condition is false      
```

```python
num = 10
if num > 0:
    print ('Positive')
else:
    print('negative')
```

```python
num = -10
if num > 0:
    print ('Positive')
else:
    print('negative')
```

```python
num = 0
if num > 0:
    print ('Positive')
elif num == 0:
    print('zero')
else:
    print('negative')
```

```python
Excel: if(condition/s, TrueValue, FalseValue)

TrueValue if condition/s else FalseValue 
#2 conditions
num = 0 
'positive' if num > 0 else 'negative'
```

```python
#3 conditions
num = 0 
'positive' if num > 0 else 'negative' if num < 0 else 'zero'
```

```python
#scenario 2
we may want to write the same code multiple times 
for this we need to use LOOP -- means iterating the same thing again and again
```

```python
2. loop [for (ranged loop) while (conditional loop)]
    1. While Loop -- we do not know how many times we have to iterate/repeat
    while --- also known as entry control LOOP -- bcz we may want to run the condition till the time this condition is true

#syntax
while condition/s:
    line1
    line2
```

```python
num = 1
while num <= 10:
    print(num) ---- #use stop button or Kernel - interrupt
    
This will keep on repeating the same true answer till the time it is intrupted or any condition to make it false
```

```python
num = 1
while num <= 10:
    print(num)
    num = num + 1  # num += 1 ----- another way to write the same thing
```

# Range

```python
# function: range()
#syntax
range(start, end, step/increament) --- complete argument and there are 3 different ways to write
range(end)  # by default it will take the start = 0, increament = 1
range(start, end) # by default the increament = 1

#end in range function goes till end -1 (only when increament is +ve)
#end in range function goes till end +1 (only when increament is -ve)
```

```python
2. loop [for (ranged loop) while (conditional loop)]
    2. for loop --- we know how many times to repeat. eg 10 times, 50 times
#syntax
for variablename in range(start, end, step/increament):
    line1
    line2
```

```python
for i in range(10):
    print(i)
```

```python
for i in range(1, 10):
    print(i)
#here we are using as membership operator
```

```python
for i in range(-3, 10):
    print(i)
```

```python
for i in range(1, 10, 2):
    print(i)
    
#it will start from 1 and increase by 2 and it will check till 10
```

```python
for i in range(2, 10, 2):
    print(i)
#it will start from 2 and increase by 2 and it will check till 10
```

```python
for i in range(2, 10, 3):
    print(i)
#it will start from 2 and increase by 3 and it will check till 10
```

```python
for i in range(1, 10, -1):
    print(i)
#this will not give any result    
```

```python
for i in range(10, 1, -1):
    print(i)
```

```python
for i in range(-1, -10, -1):
    print(i)
```
