# Missing value treatment 

```python
# How will you treat the missing values

1. You can remove a column - depending on data nad what are we trying to find
2. You can remove a row -- depending on data and what are we trying to find 
3. You can replace it with mean/average, median, mode or 0
4. You can run another model -- to get the missing value 

Example1 -if we have a data from which we are trying to find the Sales and there is a column - email with missing values
    Treating missing values -- we can remove the email column as email is not important to calculate the sales.

Example2 -if we have a data from which we are trying to find the Sales and there is a row with missing value under sales column 
    Treating missing values - we can remove the row with the missing value as it will not impact much
    
Example3 -if we have a data in which 70% of the data is missing in one column 
    Treating missing value -- we can remove that column until and unless that row is very important from business prospective
    And if we keep that we need to check by which we can replace - mean/average, median or mode

Example3 - if we have a data where 30% of the data in a column is missing.
    Treating the missing value -- if we can think of removing the 30% row then it is possible that we are removing the important
    information from other columns hence we need to take business decision.     

Example4 -if have a data in which we are trying to calculate the feedback given by our customers for one of our product. In this
    40% of our customers have given the feedback and rest have not even attempted.
    Treating missing value -- we can impute by 0 (zero) over here. 
```

# Replacing with Mean/Average, Median, Mode

```python
Mean - average (average = total/count)
Median - middle vale 
Mode - value that is most occuring 

Check the distribution of data --

1. if it is clustered towards the centre --- replace with mean/average 
2. if the data is qued/clustered on left side or right side -- replace it with median (median is middle value) after sorting the
    data in ascending order then what ever is the middle point is the middle value
        a) in case of odd numbers -- middle value
        b) in case of even numbers -- take the average of the middle values
3. if the value is a text value -- replace it with mode bcz we cannot do any calculation but we can check the value that
    is most occuring   
```

# Bins/Groups

```python
# When we have continuous variables we try to make it categorical for better understanding of data 
# when we convert continuous variable to categorical variable it is always ordinal variable whether it is group or bin
# -- So we have 2 options to do this
1. bins (equal frequency distribution table) - divide the data in such a way that the class interval is same 
Example - we have data from 0 - 100 and i want to divide the data into 10 different parts. so, we do 
 0 -  9
10 - 19
20 - 29
30 - 39
40 - 49
.. - 100
hence, here the class interval is same and is known as bins

2. groups - need not be equal interval
Example - Age -- we can convert age into bin or into groups. like crowd between 
kids 0 - 3
teen age 13 - 11
young age 20 - 27
matured people 28 - 40
and so on 

```

```python
# Signature: pd.cut(x, bins, right=True, labels=None, retbins=False, precision=3, include_lowest=False, duplicates='raise',)
# Docstring: Bin values into discrete intervals.
# x : array-like The input array to be binned. Must be 1-dimensional.
# .cut is based on the value of the records

pd.cut?
```

```python
# this has grouped the variables into 4 bins and the class interval is same which means it is bin 
# Categories (4, interval[float64]): [(70.699, 171.325] < (171.325, 271.55] < (271.55, 371.775] < (371.775, 472.0]]

pd.cut(stores.TotalSales, 4)
```

```python
# Now i want to give the class interval myself - will give min value, max value and the class interval
# min value - 0
# max value - 600
# class interval - 50

# When a bin is created it comes in (] -- which means value with ( is not included and value with ] is included
# Categories (11, interval[int64]): [(0, 50] < (50, 100] < (100, 150] < (150, 200] ... (350, 400] < (400, 450] 
# < (450, 500] < (500, 550]]
# if you want to include the right side value then write right=false in the argument

pd.cut(stores.TotalSales, range(0, 600, 50))
```

```python
# if you want to label the bins 

pd.cut(stores.TotalSales, 4, labels = ['l1', 'l2', 'l3', 'l4'])
```

```python
# how to create groups
# unequal class interval 

pd.cut(stores.TotalSales, [0, 100, 150, 180, 300, 450, stores.TotalSales.max()])
```

```python
# Now if we want to divide the data based on the count of the records then we need to use .qcut
# Signature: pd.qcut(x, q, labels=None, retbins=False, precision=3, duplicates='raise')
# Docstring: Quantile-based discretization function. Discretize variable into equal-sized buckets based on rank or based on 
# sample quantiles.
# .qcut is based on the count of the records

pd.qcut?
```

```python
pd.qcut(stores.TotalSales, 4)
```
